

**Categorical data** pose a problem as they cannot be input into the model without transforming them into numerical data first. The most suitable method for this depends on the type of data in that column:

 - **Ordinal data** are data that have a natural order to them. They are ordered or ranked from best to worst, 1st place to 4th place etc. These columns can be ranked using label encoding. eg the Kitchen Quality Column is ranked from Poor to Excellent and it would be logical to rank this with numbers 0 to 5.


 - **Cardinal data** are data which represent the number or count of something. Cardinal data can take on numerical values (such as “1” indicating male and “2” indicating female), but those numbers don’t have mathematical meaning. You couldn’t add them together, for example. The neighborhood column is an example in this dataset and the dummy method is used.
 
How the categorical features are separated has been decided by looking at the descriptions for each individual column of the dataset on the kaggle website. e.g.'Quality' columns can easily be ranked and are ordinal columns, whereas Material such as Roof Material cannot be ranked and are cardinal.

## Feature Selection and Model Testing

**Linear Regression**

- The first model to examine the relationship between fixed features and Sale Price is Linear Regression
- The model is fit with all features first - it has an R squared of 0.866 which is a good indicator - it represents the proportion of the variance of Sale Price (the Target) which is explained by the fixed features
- The RMSE is also shown below and represents how concentrated the data is around the model or in this case the line of best fit

```python
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
import statsmodels.api as sm

model = sm.OLS(y_train_fixed, X_train_fixed)
model = model.fit()
predictions = model.predict(X_train_fixed)

plt.figure(figsize=(8,6))
plt.scatter(predictions, y_train_fixed, s=30, c='r', marker='+', zorder=10)
plt.xlabel("Predicted Values - $\hat{y}$")
plt.ylabel("Actual Values - $y$")
plt.show()
```

```python
model.summary()
```

```python
train_predictions = model.predict(X_train_fixed)
test_predictions = model.predict(X_test_fixed)

from sklearn.metrics import r2_score
from sklearn import metrics

from sklearn.metrics import mean_squared_error as ms, r2_score
train_error = sqrt(ms(y_true=y_train_fixed, y_pred=train_predictions))
test_error = sqrt(ms(y_true=y_test_fixed, y_pred=test_predictions))
print("RMSE on linear model seen data: {0:.2f}.".format(train_error))
print("RMSE on linear model unseen data: {0:.2f}.".format(test_error))

r2 = metrics.r2_score(y_train_fixed, train_predictions)
print ("Cross-Predicted R2:", r2)
```


 - The features with an absolute large coefficient and a p value smaller that 0.05 are selected to see if this improves the model
- When the RMSE and R squared between the first and second model are compared the first performed better.
